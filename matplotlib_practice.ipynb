```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Changing to object-oriented method
# Create the plot.
#plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
#plt.xlabel("Date")
ax.set_xlabel("Date")
#plt.ylabel("Fare($)")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
#plt.ylim(0, 45)
ax.set_ylim(0, 45)
# Create a title.
#plt.title("PyBer Fare by Month")
ax.set_title("PyBer Fare by Month")
# Add a grid.
#plt.grid()
ax.grid()
# Add the legend.
#plt.legend()
ax.legend()
```

```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```
